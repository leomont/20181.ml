## Neuronas artificiales como unidades de activación

Recuerda que en Logistic Regression con clasificación binaria, dado un dato $\mathbf{x} \in R^n$, modelamos la probabilidad de que pertenezca a la clase 1 como:

$$P(y=1 \mid \mathbf{x};\mathbf{w}) = g(b+\mathbf{x} \cdot \mathbf{w}) \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;donde \;\;\; g(z)=\frac{1}{1+e^{-z}}$$

En este caso distinguimos el término del bias ($b$) y no asumimos que $x_0=1$. Podemos ver esto como una neurona artificial:

## Ensamblaje de neuronas artificiales

Podemos considerar una red neuronal de una capa intermedia con dos fases: 

**FASE 1**: <br/>ensamblamos varias neuronas generando una capa intermedia entre la entrada y la salida y tenemos un conjuntos de parámetros (pesos) $\mathbf{W}_1$ :


$$ \mathbf{x^T} = \left[ \begin{array}{c} x_0 & x_1 & x_2 \end{array} \right] \;\;\;\;\;\;
\mathbf{W_1} = \left[ \begin{array}{cc}
w_0^0 & w_1^0 \\
w_0^1 & w_1^1 \\
w_0^2 & w_1^2
\end{array} \right] \;\;\;\;\;\;
\mathbf{b}_1 = \left[ \begin{array}{c} b_0 & b_1 \end{array} \right]
$$

cada columna de $\mathbf{W}_1$ corresponde a los pesos de una neurona (unidad de Logistic Regression). Con datos de $n$ columnas, y $k$ neuronas en la capa intermedia, las activaciones de esta capa vienen dadas por:

$$ \mathbf{A} = \tanh(\mathbf{x^T} \cdot \mathbf{W}_1 + \mathbf{b}_1) $$

con $\mathbf{A} \in [-1,1]^{k}$, $\mathbf{x} \in \mathbb{R}^n$,  $\mathbf{W}_1 \in \mathbb{R}^{n \times k}$ y $\mathbf{b}_1 \in \mathbb{R}^k$. Observa que usamos $\tanh$ y no la función sigmoide.

**FASE 2**:<br/> Usamos Logistic Regression sobre las activaciones:

$$\hat{y} = g(\mathbf{W}_2 \cdot \mathbf{A} + b_2)$$

con $\mathbf{W}_2 \in \mathbb{R}^k$, $b_2 \in \mathbb{R}$, $\hat{y} \in [0,1]$

Ensamblando todo en una expresión, la predicción de la red neuronal pa una entrada $\mathbf{x}$ es:

$$\hat{y} = g(\mathbf{W}_2 \cdot \tanh(\mathbf{x^T} \cdot \mathbf{W}_1 + \mathbf{b}_1) + b_2)$$
